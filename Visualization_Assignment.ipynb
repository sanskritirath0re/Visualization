```python
#5.Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

# Set the random seed for reproducibility
np.random.seed(42)
# Step 1: Define the number of features and samples
num_features = 5
num_samples = 100
# Step 2: Create a mean vector (with zeros) and a covariance matrix
mean_vector = np.zeros(num_features)
# Define a covariance matrix
# The values in the covariance matrix determine the correlations between features
cov_matrix = np.array([
    [1.0, 0.8, 0.5, 0.3, 0.1],
    [0.8, 1.0, 0.6, 0.4, 0.2],
    [0.5, 0.6, 1.0, 0.7, 0.3],
    [0.3, 0.4, 0.7, 1.0, 0.5],
    [0.1, 0.2, 0.3, 0.5, 1.0]
])
# Step 3: Generate samples from a multivariate normal distribution
data = np.random.multivariate_normal(mean_vector, cov_matrix, num_samples)
# Step 4: Convert the samples to a pandas DataFrame
columns = ['Feature_1', 'Feature_2', 'Feature_3', 'Feature_4', 'Feature_5']
df = pd.DataFrame(data, columns=columns)
# Calculate the correlation matrix
correlation_matrix = df.corr()
# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
#3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)
pivot_table = df.pivot_table(values='Sales', index='Day', columns='Month', aggfunc='sum')
# Generate the heatmap using the pivot table
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, annot=True, fmt='.1f', cmap='coolwarm', linewidths=.5)
plt.title('Sales Variation Across Different Months and Days')
plt.xlabel('Month')
plt.ylabel('Day')
plt.show()
```

```python
#4. Using the given x and y data, generate a 3D surface plot to visualize the function z=sin((x*2+y*2)^)
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)
# Pivot the DataFrame to create a 2D array for the heatmap
pivot_table = df.pivot('Y', 'X', 'Z')

# Plot the heatmap using seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(pivot_table, cmap='viridis', cbar=True)
plt.title('Heatmap of $z = \sin(\sqrt{x^2 + y^2})$')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
